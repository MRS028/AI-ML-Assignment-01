```python
import numpy as np
import pandas as pd

np.random.seed(42)

ids = np.arange(1, 11)
ages = np.random.randint(18, 60, 10)
salaries = np.random.randint(30000, 90000, 10)
departments = np.array(["HR", "IT", "Finance", "IT", "HR", "Sales", "Finance", "IT", "Sales", "HR"])

DF = pd.DataFrame({
"id": ids,
"age": ages,
"salary": salaries,
"dept": departments
})

DF.to_csv("employees.csv", index=False)
print("Sample Data Created and Saved as employees.csv")
```

**Answer to te question No:Q1**

```python
ages_array = np.array(ages)
salaries_array = np.array(salaries)

print("Ages Array:", ages_array)
print("Salaries Array:", salaries_array)

print("\nAges Array Properties:")
print("dtype:", ages_array.dtype)
print("ndim:", ages_array.ndim)
print("shape:", ages_array.shape)
print("size:", ages_array.size)

print("\nSalaries Array Properties:")
print("dtype:", salaries_array.dtype)
print("ndim:", salaries_array.ndim)
print("shape:", salaries_array.shape)
print("size:", salaries_array.size)
```

**Answer to te question No:Q2**

```python
max_salary = np.max(salaries_array)
min_salary = np.min(salaries_array)
avg_salary = np.mean(salaries_array)
avg_age = np.mean(ages_array)

print("Highest Salary:", max_salary)
print("Lowest Salary:", min_salary)
print("Average Salary:", round(avg_salary, 2))
print("Average Age:", round(avg_age, 2))
```

**Answer to te question No:Q3**

```python
age_condition = ages_array > 30
ages_above_30 = ages_array[age_condition]
count_above_30 = np.sum(age_condition)

print("Ages greater than 30:", ages_above_30)
print("Number of employees older than 30:", count_above_30)
```

**Answer to te question No:Q4**

```python
updated_ages = ages_array + 5

print("Original Ages Array:", ages_array)
print("Updated Ages Array (+5 years):", updated_ages)
print("Original array unchanged:", ages_array)
```

**Answer to te question No:Q5**

```python
total_salary_expense = np.sum(salaries_array)
salary_difference = np.ptp(salaries_array)

print("Total Salary Expense(sum of all salaries):", total_salary_expense)
print("Difference between Max & Min Salary:", salary_difference)
```
